```python
import json
import geojson 
import geopandas
import pandas as pd
import numpy as np
import os
```

```python
#extract calibration points, export as list for lmdpy
#order of input list matters for lmd callibration
def get_calib_points(list_of_calibpoint_names, df):
    #create shape list
    pointlist = []
    for point_name in list_of_calibpoint_names:
        pointlist.append(df.loc[df['name'] == point_name, 'geometry'].values[0])
    #create coordenate list
    listarray = []
    for point in pointlist:
        listarray.append([point.x, point.y])
    nparray = np.array(listarray)
    return(nparray)

#returns dataframe with only polygon objects
def remove_non_polygons(df):
    #convert entire df to text
    df_str = df.astype(str)
    #filter original df by finding the word polygon, poly_df created
    poly_df = df[df_str['geometry'].str.contains("POLYGON")]
    poly_df = poly_df.reset_index(drop=True)
    return(poly_df)

#creates new column for coordenates in a list of list format
#assumes only polygons in dataframe
def replace_coords(df):
    df['lol'] = np.nan
    df['lol'] = df['lol'].astype('object')
    
    for i in df.index:
        #get geometry object for row i
        geom=df.at[i, 'geometry']
        #list the coordenate points as tuples
        tmp = list(geom.exterior.coords)
        #transform list of tuples to list of lists
        tmp_lol = [list(i) for i in tmp]
        #add list of list to lol column
        df.at[i,'lol'] = tmp_lol
    
    return(df)

#creates a column for polygon type
#assumes only polygons in dataframe
def get_contour_type(df):
    #create empty column
    df['Name'] = np.nan
    #fill empty column with whatever is inside dictionary name entry
    for i in df.index:
        tmp = df.classification[i].get('name')
        df.at[i,'Name'] = tmp
    return(df)
```

### intructions for users

1. Open Qupath, create new project, add images, modify histogram for every channel
2. Click Annotations

3. Set Classes for your cell types of interest.

    3.1 You can add classes by right clicking in the class menu and clicking add class.
    
    3.15 You can remove the default ones by shift clicking them and then clicking the three dot button below -> Add/Remove -> Remove class
    
    3.2 You can add as many classes as you want.
    
    3.3 The class names should be text, without any numbers, symbols, or spaces, use underscore if needed ("_"). 
    
    3.4 Class names are going to be used later to select which cells go to which well, so remember to write them down.
    
    3.5 There is no need to create classes for replicates, the separation is done later, unless you don't trust the random selection.
    
    3.6 I am working on adding celltype and spatial information in classnames. 
    
    3.7 If you save your project classes will also be saved with it. 

4. Calibration points. You need a minimum of three calibration points. 

    4.1 To create a calibration point:
    
        4.1.1 Go to the annotation menu.
        4.1.2 Click the 3 circle logo in the menu
        
        4.1.3 A little menu pops. Click Add. Click Edit. 
        4.1.4 Give your calibration point a name and click apply
        4.1.5 Click in the image, only once. Click Add for the next point.

Some notes regarding calibration points:

Naming of calibration points: No spaces or symbols are allowed. Only characters, numbers, and underscores are allowed. I recommned calib1, calib2, calib3, and so on. Please write down their names for later.

The larger the distance between calibration points and contours to be cut, the greater the probability of contour distortion. However, using many calibration points is complicated because you must choose which contours to use for which calibration points. The order of the calibration points is important. Make sure they are numbered, and write down the order you will use. It has to match between the contour creating steps and the lmd microscope.

Jose's recommendation: If you can select everything inside three calibration points do it. 3mm^2 is the maximum range I would recommend for 3 calibration points. How to check? In Qupath select the Line tool.

And draw lines between your three calibration points. Optimally all your contours are inside the triangle you just made. This is not required though, the purple circle in the image below is also good enough. 



5. Adding shapes:

    5.1 You can select any contour tool: square, circle, polygon, and the paintbrush (looks like a cell during mitosis). Create any annotation by clicking and draging. You can always edit the annotations later. You can delete area by using ALT in windows, not sure in Mac. 
    
    5.2 After you are happy with your annotation, you have to click in the class you want to label with, and click "Set class".

6. Export:

When you are done with both calibration points and contours to be cut, go to the File menu and click on "Export Objects as GeoJSON". An Export menu will pop.

Make sure the first dropdown menu says -> all objects. and check Pretty JSON. click OK.


Save your file somewhere, give it a good name :)
